### Build the `text-to-speech` pipeline using ðŸ¤— Transformers Library

````python
from transformers.utils import logging

logging.set_verbosity_error()
````

````python
from transformers import pipeline

narrator = pipeline("text-to-speech",
                    model="kakao-enterprise/vits-ljs")
````

_Note_: Find more information about the model 'kakao-enterprise/vits-ljs', [here](https://huggingface.co/kakao-enterprise/vits-ljs).

````python
# Set a text sample
text = """Assam is a state in northeastern India, situated south of the eastern Himalayas along the Brahmaputra and Barak River valleys.
Assam covers an area of 78,438 square kilometers (30,285 sq mi).
It is known for its tea, large and old petroleum resources, and biodiversity.
Assam has a population of more than 35 million, making it the most populous state in Northeast India and the 17th most populous state in India."""
````

````python
# Send the text to the pipeline
narrated_text = narrator(text)
````

````python
# Listen to the text converted to audio
from IPython.display import Audio as IPythonAudio

IPythonAudio(narrated_text["audio"][0],
             rate=narrated_text["sampling_rate"])
````
